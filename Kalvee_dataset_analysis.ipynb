```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold

# Load the dataset
data =  pd.read_csv("Model_implementation_final.csv")
#data = np.loadtxt("Model_implementation_final.csv", delimiter=",")



# Split the dataset into k folds
kf = KFold(n_splits=5)

print(kf)
clf = RandomForestClassifier(n_estimators=100)

# Perform k-fold cross-validation
scores = []
for train_index, test_index in kf.split(data):
    X_train, X_test = x.iloc[:,train_index[0]:train_index[-1]],x.iloc[:,test_index[0]:test_index[-1]]
    y_train, y_test = y[train_index[0]:train_index[-1]], y[test_index[0]:test_index[-1]]
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    scores.append(score)

# Print the average score
print("Average score:", np.mean(scores))

```
